# **ResNet50**

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import LeakyReLU

leaky_relu_alpha = 0.1

# Initialize the Pretrained Model
feature_extractor = ResNet50(weights='imagenet',
                             input_shape=(80,100, 3),
                             include_top=False)

# Set this parameter to make sure it's not being trained
feature_extractor.trainable = False

# Set the input layer
input_ = tf.keras.Input(shape=(80,100, 3))

# Set the feature extractor layer
x = feature_extractor(input_, training=False)

# Set the pooling layer
# x = tf.keras.layers.LeakyReLU(alpha=0.2)
# x.add(Conv2D(32, kernel_size=(3, 3), input_shape=input_))
# x.add(LeakyReLU(alpha=leaky_relu_alpha))
x = tf.keras.layers.GlobalAveragePooling2D()(x)
tf.keras.layers.LeakyReLU(alpha=0.3)
# x = tf.keras.layers.Dropout(0.3)(x)


x = tf.keras.layers.Dense(512)(x)
tf.keras.layers.LeakyReLU(alpha=0.3)

x = tf.keras.layers.Dropout(0.8)(x)
x = tf.keras.layers.Dense(128, activation='leaky_relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
# x = tf.keras.layers.Dense(128, activation='relu')(x)
# x = tf.keras.layers.Dropout(0.6)(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)

# Set the final layer with sigmoid activation function
output_ = tf.keras.layers.Dense(19, activation='softmax')(x)

# Create the new model object
model = tf.keras.Model(input_, output_)

# Compile it
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

# Print The Summary of The Model
model.summary()
```

```python
model.fit(X_train,Y_train, epochs = 100 ,validation_split=0.1)
#model.fit(X_train,Y_train, epochs=20, validation_data=X_test)
```

# **Densenet201**

```python
shape=(100,80,3)
input_tensor=keras.Input(shape=shape)
base_model=tf.keras.applications.densenet.DenseNet201(input_tensor=input_tensor,weights='imagenet',include_top=False)
# base_model.trainable = False
avg=keras.layers.AveragePooling2D(2,padding='valid')(base_model.output)
depth=keras.layers.DepthwiseConv2D(128, (3, 3), padding="same", activation="relu", depth_multiplier=2)(avg)
#depth=keras.layers.DepthwiseConv2D(64, (3, 3), padding="same", activation="relu", depth_multiplier=2)(depth)
flat=keras.layers.Flatten()(depth)
dp=keras.layers.Dropout(0.5)(flat)
# model_alexnet.add(BatchNormalization())
preds=keras.layers.Dense(19,activation='softmax',
                          kernel_initializer=keras.initializers.RandomNormal(mean=0.0,stddev=0.01),
                          bias_initializer=keras.initializers.Zeros(),)(dp)
model_den=keras.Model(inputs=base_model.input, outputs=preds)

# Compile it
model_den.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])


model_den.summary()
```

```python
from keras.callbacks import ModelCheckpoint
cp_checkpoint = ModelCheckpoint("my_model_den", save_best_only=True, verbose=1)

model_den.fit(X_train, Y_train,
          validation_split =0.1,
          epochs=400,
          callbacks=[cp_checkpoint])
```

```python
preds_den= model_den.predict(X_test)
all_preds_den=[] # this list will contain all the predicted labels of the DenseNet Network
for i in preds_den:
  x=list(i).index(max(i))
  all_preds_den.append(x)
```
